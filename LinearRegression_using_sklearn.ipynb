Linear Regression using sklearn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('E:/Datasets/Weather.csv')
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset.head()
```

```python
dataset.plot(x='MinTemp', y='MaxTemp', style='o')
plt.title('MinTemp vs MaxTemp')
plt.xlabel('MinTemp')
plt.ylabel('MaxTemp')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.distplot(dataset['MaxTemp'])
```

```python
x= dataset['MinTemp'].values.reshape(-1,1)
y= dataset['MaxTemp'].values.reshape(-1,1)
```

```python
#splitting the data into train and test data
#80% data in training and rest in test set
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
regressor=LinearRegression()
regressor.fit(x_train, y_train) #training the data
```

```python
print("Regressor_Intercept=",regressor.intercept_)
print("Regressor_Coefficient=", regressor.coef_)
```

```python
y_pred = regressor.predict(x_test)
```

```python
#creating a dataframe

df = pd.DataFrame({'Actual':y_test.flatten(), 'Predicted':y_pred.flatten()})
```

```python
df
```

```python
df1= df.head(25) #transferring 25 top entries of df into df1
df1.plot(kind='bar', figsize=(32,20))
plt.grid(which='major',linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle='-', linewidth='0.5',color='black')
```

```python
#plotting the predictions
plt.scatter(x_test, y_test, color='orange')
plt.plot(x_test, y_pred, color="black", linewidth=2)
```

Printing "Mean absolute error", "Mean Squared Error", "Root Mean Squared Error"

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
